PROBLEM 1: Display 'Manager' and 'Grand Total Sales', for sales done by the sales agents reporting these managers

```python
#solution code here
df_salesteams.head()
df = df_salespipeline[df_salespipeline['deal_stage'] == 'Won']
df = df[['sales_agent','close_value']]
x = df.groupby(['sales_agent'], as_index = False, sort=False).sum()
final = pd.merge(df_salesteams,x,how='inner')
final = final.drop(['sales_agent','regional_office'],axis=1)
ans = final.groupby(['manager'],as_index=False,sort=False).sum()
ans = ans.rename(columns={'manager': 'Manager','close_value' : 'Grand Total Sales'})
ans
```

PROBLEM 2: Display 'Sales Agents' and 'Sales' for those sales where product sold at profit

```python
#solution code here
df_products.head()
a = df_salespipeline[df_salespipeline['deal_stage'] == 'Won']
a = a[['sales_agent','product','close_value']]
y = pd.merge(df_products,a,how='inner')
y = y[(y['close_value'] > y['sales_price'])]
y.reset_index(drop=True,inplace=True)
y['profit'] = (y['close_value'] - y['sales_price'])
y = y.drop(['sales_price','close_value'],axis=1)
ans2 = y.groupby(['sales_agent'],as_index=False,sort=False).count()
ans2['sales'] = ans2['profit']
ans2.drop(['product','profit'],axis=1,inplace=True)
# The sales columns denotes the count of sales done by agent which resulted in profit
ans2
```

PROBLEM 3: Display the 'Opportunity ID' and 'Days Taken to Close', for opportunities those got closed within a month

```python
#solution code here
df_salespipeline.head()
b = df_salespipeline[(df_salespipeline['deal_stage'] == 'Won') | (df_salespipeline['deal_stage'] == 'Lost')]
b = b[['opportunity_id','created_on','close_date']]
b.reset_index(drop=True,inplace=True)
b['created_on'] = pd.to_datetime(b['created_on'])
b['close_date'] = pd.to_datetime(b['close_date'])
b['created_year'] = b['created_on'].dt.year
b['closed_year'] = b['close_date'].dt.year
b = b[(b['created_year']) == (b['closed_year'])]
from datetime import date
b['Days Taken to Close'] = b['close_date'] - b['created_on']
b['Days Taken to Close'] = b['Days Taken to Close'] / np.timedelta64(1,'D')
b['Days Taken to Close'] = b['Days Taken to Close'].astype(int)
b['months Taken to Close'] = b['close_date'] - b['created_on']
b['months Taken to Close'] = b['months Taken to Close'] / np.timedelta64(1,'M')
b = b[b['months Taken to Close'] <= 1]
b.reset_index(drop=True,inplace=True)
ans3 = b[['opportunity_id','Days Taken to Close']]
ans3
```

PROBLEM 9: Which sales agent(s) never lost a deal. Display as a dictionary {sales agent:sales}

```python
#solution code here
df_salespipeline.head()
data = df_salespipeline[['sales_agent','deal_stage']]
data = data[(data['deal_stage'] == 'Lost') | (data['deal_stage'] == 'Won')]
val = data.groupby(['sales_agent'],as_index=False,sort=True)
array = []
for x,y in val:
    count = 0
    for i in y['deal_stage']:
        if (i == 'Lost'):
            count = count + 1
    if (count != 0):
        array.append(x)
if (len(array) == len(df_salesteams['sales_agent'])):
    print('There is no such sales agent who never lost a deal')
```
